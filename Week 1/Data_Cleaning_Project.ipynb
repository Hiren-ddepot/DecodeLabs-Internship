```python
import pandas as pd

# Load dataset
file_path = r"C:\Users\onyer\Downloads\Dataset for Data Analytics.xlsx"
df = pd.read_excel(file_path)
```

```python
# Dataset Overview
print(df.head(10))
print("\nThe size of the dataset:")
print("Columns:", df.shape[1])
print("Rows:", df.shape[0])
```

```python
# Missing Values
print("Missing values before cleaning\n", df.isnull().sum())
```

```python
# Replace missing CuouponCode with "No Coupon"
df["CouponCode"] = df["CouponCode"].fillna("NO COUPON")
print("Missing values after cleaning\n", df.isnull().sum())
```

```python
# Duplicated values 
print("Duplicated Rows found:", df.duplicated().sum())
```

```python
# Check for duplicated CustomerID
print("Duplicated OrderID found:", df["OrderID"].duplicated().sum())
```

```python
# Check for data types
print("Data types\n", df.dtypes)
```

```python
# Converting Date to correct datetime format
df["Date"] = df["Date"].astype("datetime64[ns]")
print("Data types after conversion\n", df.dtypes)

invalid_dates = df["Date"].isnull().sum()
print("\ninvalid_dates:", invalid_dates)
```

```python
# Numerical columns validation
df.select_dtypes(include=["number"]).dtypes
```

```python
df[["Quantity", "UnitPrice", "ItemsInCart", "TotalPrice"]].isna().sum()
```

```python
# Checked for trailing whitespace
df = df.apply(lambda col: col.str.strip() if col.dtype == "object" else col)

df["Product"].head()
```

```python
# check for outlier
print(df.describe())
```

```python
# Copy of the cleaned dataset
df_clean = df.copy()
print(df_clean.head())
```

```python
df_clean.to_excel("Week-01-Cleaned-Dataset.xlsx", index=False)
print("Save successful as Week-01-Cleaned-Dataset.xlsx")
```
